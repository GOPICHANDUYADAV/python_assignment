1. Factorial calculator:

```python
def factorial(n):
    if n == 0:
        return 1

    result = 1
    for i in range(1, n+1):
        result *= i

    return result


print(factorial(5))
```

2. compound intrest calculator

```python
def compound_interest(principal, rate, time, n):
    r = rate / 100
    amount = principal * (1 + r/n) ** (n*time)
    return round(amount,2)


print(compound_interest(1000,5,2,12))
```

3. Longest Word Finder:

```python
def longest_word(sentence):
    sentence = sentence.replace(".", "").replace(",", "").replace("!", "")
    words = sentence.split()

    longest = ""

    for word in words:
        if len(word) > len(longest):
            longest = word

    return longest


print(longest_word("The quick brown fox jumps over the lazy dog"))
```

4. Anagram checker

```python
def are_anagrams(str1, str2):

    str1 = str1.lower().replace(" ","")
    str2 = str2.lower().replace(" ","")

    if sorted(str1) == sorted(str2):
        return True
    else:
        return False


print(are_anagrams("listen","silent"))
```

5. vowels and consonents counter:

```python
def count_vowels_consonants(text):

    vowels = "aeiouAEIOU"
    v = 0
    c = 0

    for ch in text:
        if ch.isalpha():
            if ch in vowels:
                v += 1
            else:
                c += 1

    return (v,c)


print(count_vowels_consonants("gopi chandu"))
```

6. List Intersection Finder:

```python
def intersection(list1, list2):

    result = []

    for i in list1:
        if i in list2 and i not in result:
            result.append(i)

    return result


print(intersection([1,2,3,4,5],[4,5,6,7]))
```

7. Fibonacci series Generator:

```python
def fibonacci(n):

    if n == 0:
        return []
    if n == 1:
        return [0]

    fib = [0,1]

    for i in range(2,n):
        fib.append(fib[i-1] + fib[i-2])

    return fib


print(fibonacci(10))
```

8. Expected Output

```python
def char_frequency(text):

    freq = {}

    for ch in text.lower():
        if ch.isalpha():

            if ch in freq:
                freq[ch] += 1
            else:
                freq[ch] = 1

    return freq


print(char_frequency("Hello World"))
```

9. Simple Calculator Function:


```python
def calculator(num1, num2, op):

    if op == "+":
        return num1 + num2

    elif op == "-":
        return num1 - num2

    elif op == "*":
        return num1 * num2

    elif op == "/":
        if num2 == 0:
            return "Error: Division by zero"
        return num1 / num2

    elif op == "%":
        return num1 % num2

    elif op == "**":
        return num1 ** num2

    else:
        return "Invalid operator"


print(calculator(10,5,"+"))
```

10. Group Anagrams

```python
def group_anagrams(words):

    groups = {}

    for word in words:
        key = ''.join(sorted(word.lower()))

        if key not in groups:
            groups[key] = []

        groups[key].append(word)

    return list(groups.values())


print(group_anagrams(["eat","tea","tan","ate","nat","bat"]))
```

```python
print(group_anagrams(["abc","def","ghi"]))
```

11. Longest Unique Substring

```python
def longest_unique_substring(s):

    longest = 0

    for i in range(len(s)):

        current = ""

        for j in range(i,len(s)):

            if s[j] not in current:
                current += s[j]
                longest = max(longest,len(current))
            else:
                break

    return longest


print(longest_unique_substring("abcabcbb"))
```

```python
print(longest_unique_substring("bbbbbbb"))
```

12. Student Grade Calculator:

```python
def get_grade(p):

    if p >= 90:
        return "A+"
    elif p >= 80:
        return "A"
    elif p >= 70:
        return "B+"
    elif p >= 60:
        return "B"
    elif p >= 50:
        return "C"
    elif p >= 40:
        return "D"
    else:
        return "F"


def calculate_grades(marks_dict):

    result = {}

    for student in marks_dict:

        marks = marks_dict[student].values()

        total = sum(marks)
        subjects = len(marks)

        percentage = round(total/subjects,1)

        grade = get_grade(percentage)

        if percentage >= 40:
            status = "Pass"
        else:
            status = "Fail"

        result[student] = {
            "percentage":percentage,
            "grade":grade,
            "status":status
        }

    return result


marks = {
"Alice":{"Math":95,"Science":88,"English":92},
"Bob":{"Math":45,"Science":38,"English":52},
"Charlie":{"Math":75,"Science":82,"English":68}
}

print(calculate_grades(marks))
```
